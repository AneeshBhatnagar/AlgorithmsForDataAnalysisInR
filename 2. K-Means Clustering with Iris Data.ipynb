## Splitting Data Frame
Here, we are just splitting the data frame into x and y, where x contains all the 4 features of the table and y contains the class (output) of the data frame.

```r
x = iris[,-5]
y = iris$Species
```

```r
kc <- kmeans(x,3)
kc
```

## Comparing the result
Comparing the result of the actual outputs with the ones computed using k means. It is done using the table function in R.

```r
kc$centers
```

```r
table(y,kc$cluster)
```

## Plotting The Output
Here, we are just plotting the output of the clustering with just the sepal length and the sepal width, colored according to their cluster number, 1,2, or 3.

```r
plot(x[c("Sepal.Length", "Sepal.Width")], col=kc$cluster)
points(kc$centers[,c("Sepal.Length", "Sepal.Width")], col=1:3, pch=23, cex=3)
```
